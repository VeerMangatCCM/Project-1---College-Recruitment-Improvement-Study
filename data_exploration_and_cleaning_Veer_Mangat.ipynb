## 3. Data Cleaning
### 3.1 Rename columns to lowercase with underscores

#### In this section, I used the suggestion to turn them all lowercase and have underscores replace spaces to standardize the naming conventions. I used this w3 school link (https://www.w3schools.com/python/pandas/ref_df_replace.asp), along with this StackOverflow post (https://stackoverflow.com/questions/42750551/converting-strings-to-a-lower-case-in-pandas)

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

### 3.2 Keep only relevant columns
#### I start off by listing out all of the columns, since now they are renamed. I decided that my questions can be answered by only keeping questions that relate to motivations, the degree/program they're enrolled in, demographic questions, and which topic interests non-majors the most. However, if the column doesn't have data, that is for a reason. Therefore, I kept ALL columns that met my criteria of being part of that question set. I chose to save the columns that met the criteria into a new dataset, instead of dropping all of the other ones  I felt this was more efficient.
#### I found these links helpful in my work (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.copy.html), (https://www.statology.org/pandas-keep-columns/)

```python
df.columns.tolist()
```

```python
keep_columns = ['timestamp', 'what_degree_program_are_you_currently_enrolled_in?','gender','race/ethnicity','age_'
] + [c for c in df.columns if 'what_motivated_you_to_seek_a_computing_class' in c] + [c for c in df.columns if 'if_you_answered_that_you_were_interested_in_taking_more_computing_classes' in c]

df_clean = df[keep_columns].copy()
df_clean.head()
```

### 3.3 Save cleaned dataset
#### (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html)

```python
df_clean.to_csv('cleaned_survey.csv')
```
